```python
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('github_source/sales_by_employee.csv', parse_dates=['Date'])
ds
```

```python
ds.pivot_table(index='Date', values=['Revenue', 'Expenses'], aggfunc='sum')
```

```python
ds.pivot_table(index='Date', values=['Revenue', 'Expenses'], aggfunc='sum', columns='Name')
```

```python
ds.pivot_table(index='Date', values=['Revenue', 'Expenses'], aggfunc='sum', columns='Name', margins=True, margins_name='Total')
```

```python
ds.pivot_table(index='Date', values=['Revenue', 'Expenses'], aggfunc='sum', columns='Name', margins=True, margins_name='Total', fill_value=0)
```

```python
ds.pivot_table(index='Date', values='Revenue', aggfunc=['count','sum'], columns='Name')
```

```python
ds.pivot_table(index='Date', values=['Revenue', 'Expenses'], aggfunc={'Revenue':'count', 'Expenses':'sum'}, columns='Name')
```

```python
#الان در زیر ما مالتی ایندکس داریم
ds.pivot_table(index=['Date', 'Name'], values='Revenue')
```

```python
# ستونها را ریخته تو ریدفاها،   اوردن ستونها به عنوان ایندکس ، و همون بحث مالتی انیدکس که در بالا هم اومده 
new = ds.pivot_table(index='Date', values='Revenue', aggfunc='sum', columns='Name')
new.stack()
```

```python
#ردیف ها را ریخته تو ستونها ها
new = ds.pivot_table(index=['Date', 'Name'], values='Revenue')
new.unstack()
```
