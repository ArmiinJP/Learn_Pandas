```python
import pandas as pd
import numpy as np
```

```python
ep = pd.read_csv('github_source/employees.csv', parse_dates=['Start Date'])
ep
```

```python
# گفتیم loc , iloc براساس ایندکس میومد فیلتر میکرد
# اگه نخوایم بر اساس ستون ایندکس باشه و بر اساس هر ستونی خواستیم اوکی کنیم چه کنیم؟

ep[ep['First Name'] == 'Maria'] # اونایی که اسمشون ماریاست
ep[ep['Team'] != 'Finance'] #اونایی که تو فایننس نیستند
ep[ep['Mgmt'] != True] #اونایی که مدیر دارند
```

```python
hSalary = ep['Salary'] > 100000
ep[hSalary].sort_values('Salary', ascending=False)
```

```python
# برای چند شرطی هم از & , | میشه استفاده کرد
isFemale = ep['Gender'] == 'Female'
#isFemale = ep['Gender'].eq('Female')
isFinance = ep['Team'] == 'Finance'
ep[isFemale & isFinance].head(3)
#ep[~isFemale & isFinance]
```

```python
# وقتی میخوایم از چند شرط داخل یک ستون استفاده کنیم به جای اینکه سه تا بزاریم و اند بزاریم بینشون میشه از روش زیر استفاده کرد
ep[ep['Team'].isin(['Marketing', 'Finance'])]
```

```python
# برای یک محدوده هم به جای اینکه دو تا شرط بزاریم میشه از روش زیر استفاده کرد
ep[ep['Salary'].between(80000, 90000)].sort_values('Salary')
```

```python
ep[ep['Start Date'].between(left='1980-01-01', right='1990-01-01')]
ep[ep['First Name'].between(left='R', right='U')].head(3)
```

```python
ep[ep['Team'].isnull()].head(2)
```

```python
#ep[~ep['Team'].isnull()].head(2)
ep[ep['Team'].notnull()].head(2)
```

```python
#تعداد ردیف های خروجی
ep.shape[0]
# این دراپ هر جا که نن باشه در هر فیلدی ، سطر اونو حذف میکنه
ep.dropna().shape[0]
#len(ep.dropna())
```

```python
#اون ردیفهایی که تمامی فیلدهاش نن هست را حذف میکنه
ep.dropna(how='all').shape[0]
```

```python
#بین این دو ستون هر ردیفی که این دو مقدارش نن باشه حذف میکنه
ep.dropna(subset=['Gender', 'Team']).shape[0]
```

```python
#هر ردیف حداقل چهار مقدارش پر شده باشه و نن نباشه اگه کمتر بود حذف شه
ep.dropna(thresh=4).shape[0]
```

```python
#حذف داپلیکیتد ها و حفظ کردن اولین باری که اومدند
ep[~ep['Team'].duplicated()]
```

```python
#اگه میخوای به جای اولین باری که اومدن، اخرین باری که اومدن و نگه داره
ep[~ep['Team'].duplicated(keep='last')]
```

```python
#این متود میاد ردیف هایی که مثل هم هستند را حذفشون میکنه
e1 = ep['Gender'] == 'Male'
e2 = ep['Team'] == 'Marketing'
ep[e1 & e2][['Gender', 'Team']].drop_duplicates()
```
