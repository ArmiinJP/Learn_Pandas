```python
import pandas as pd
```

```python
food_data = {
    "Item": ["banana", "cucamber", "orange", "tomato", "watermelon"],
    "Type": ["fruit", "vegetable", "fruit", "vegetable", "fruit"],
    "Price": [0.99, 1.25, 0.25, 0.33, 3.00]
}
supermarket = pd.DataFrame(data=food_data)

g = supermarket.groupby(by='Type')
g.mean(numeric_only=True)
```

```python
g.get_group("fruit").mean(numeric_only=True)
```

```python
ft = pd.read_csv("github_source/fortune1000.csv")
sectors = ft.groupby(by="Sector")
sectors.size()
```

```python
len(sectors)
```

```python
sectors.groups
sectors.first()
sectors.nth(3)
```

```python
sectors['Profits'].sum(numeric_only=True)
```

```python
sectors.agg({
    'Revenues': 'sum',
    'Profits': 'mean',
    'Employees': 'min'
})
```

```python
def get_largest_each_group(df):
    return df.nlargest(1, 'Profits')

sectors.apply(get_largest_each_group)
```
